## EDA

```python
import pickle 
import pandas as pd 
from plotnine import ggplot, aes, geom_bar, geom_point, geom_histogram, geom_boxplot
```

```python
infile = open('data2.pkl','rb')
df = pickle.load(infile)
infile.close()
```

```python
df1 = df.copy()
```

```python
df1.dropna(inplace =True)
```

```python
ggplot(df1, aes(x = 'release_speed', fill = 'pitch_name')) + geom_histogram()
```

4-seam fastballs stand out for being faster then other pitches. 

```python
ggplot(df1, aes(x = 'factor(pitch_type)', y = 'release_speed')) + geom_boxplot()
```
